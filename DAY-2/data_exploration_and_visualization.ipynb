```python
# -----------------------------
# Step 2: Import Required Libraries
# -----------------------------

# Numerical computations (handles arrays, math operations)
import numpy as np

# Data handling & analysis (used for DataFrames, tables, cleaning)
import pandas as pd

# Visualization libraries
import matplotlib.pyplot as plt    # Basic graphs
import seaborn as sns              # Advanced and pretty graphs

# Dataset for practice (Iris dataset)
from sklearn.datasets import load_iris

# Set default plot style
sns.set(style="whitegrid")

print("All libraries imported successfully!")
```

```python
# -----------------------------
# Step 3: Load the Iris Dataset
# -----------------------------

# Load the dataset from sklearn
iris = load_iris()

# Convert to a DataFrame for easy analysis
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target column (flower species)
data['target'] = iris.target

# Display first 5 rows
data.head()
```

```python
# -----------------------------
# Step 4: Basic Data Exploration
# -----------------------------

# Shape of dataset (rows, columns)
print("Shape of dataset:", data.shape)

print("\nColumn Names:")
print(data.columns)

print("\nData Types:")
print(data.dtypes)

print("\nSummary Statistics:")
print(data.describe())
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target

df.head()
```

```python
sns.pairplot(df, hue="target")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, linewidths=0.5)
plt.show()
```

```python
df.hist(figsize=(10, 6))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.show()
```

```python
# Check for missing values in each column
print("Missing values per column:")
print(df.isnull().sum())
```
