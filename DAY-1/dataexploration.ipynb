```python
# Step 2: Import necessary libraries

# For data handling
import numpy as np
import pandas as pd

# For data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# For loading sample datasets
from sklearn.datasets import load_iris

# For nicer-looking plots
sns.set(style="whitegrid")

print("✅ Libraries imported successfully!")
```

```python
# Step 3: Load the Iris dataset
iris = load_iris()

# Convert it into a Pandas DataFrame for easier handling
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target column (the flower species)
data['target'] = iris.target

# Display the first 5 rows
data.head()
```

```python

# Step 4: Explore basic information about the dataset

print("Shape of dataset:", data.shape)   # Rows and columns
print("\nData types:\n", data.dtypes)    # Type of each column
print("\nSummary statistics:\n", data.describe())  # Mean, min, max, etc.
```

```python
# Step 5: Check for missing values

print("Missing values in each column:\n")
print(data.isnull().sum())
```

```python
# Step 6.1: Histogram of all features
data.hist(figsize=(10,8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
# Step 6.2: Pairplot of features colored by target
sns.pairplot(data, hue='target', palette='bright')
plt.suptitle("Pairplot of Iris Features", y=1.02, fontsize=16)
plt.show()
```

```python
# Step 6.3: Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Iris Features", fontsize=16)
plt.show()
```
